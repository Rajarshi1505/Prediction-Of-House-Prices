```python
#we have to replace dates with more meaningful quatities viz. day, month, year
dates=list(df['date'])
dates
```

```python
day=[]
month=[]
year=[]
dates[0].split(' ')[0].split('-')
```

```python
for date in dates:
    x=list(map(int, date.split(' ')[0].split('-')))
    year.append(x[0])
    month.append(x[1])
    day.append(x[2])
```

```python
print(len(year))
print(len(month))
print(len(day))
```

```python
df['year']=year
df['month']=month
df['day']=day
df=df.drop(columns='date')
df.head()
```

```python
df['country'].value_counts()
```

We remove the 'country' column as there is only one type of value

```python
df=df.drop(columns='country')
df.head()
```

We now replace the columns 'yr_built' and 'yr_renovated' with more meaningful values viz. 'age' and 'time_since_last_renovation' 

```python
df['yr_built']=df['year']-df['yr_built']
df['yr_renovated']=df['year']-df['yr_renovated']
df.rename(columns={'yr_built':'age', 'yr_renovated':'time_since_last_renovation'},inplace=True)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
df=shuffle(df)
x=df.drop(columns='price')
y=df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_test.shape)
```

```python
import statsmodels.api as sm
model=sm.OLS(y,x).fit()
```

```python
print(model.summary())
```

```python
pred=model.predict(x_test)
comptb=pd.DataFrame(list(zip(pred, y_test)), columns=['pred', 'y_test'])
comptb
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(y_test, pred))
print(rmse)
```
